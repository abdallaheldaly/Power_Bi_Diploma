# why array can store data greater than other data structure ?

In a list, each element is stored with additional information about the element, such as its type and size. This additional information takes up space in memory, and can make lists less efficient for storing large amounts of data.

On the other hand, arrays store their elements in a 
###### contiguous block of memory, 
and do not have the overhead of additional information about each element. This makes arrays more efficient for storing large amounts of data, because they use less memory and can be accessed more quickly.

Contiguous blocks of memory are often used to store data structures such as arrays, because they allow elements to be accessed quickly and efficiently. When an array is stored in a contiguous block of memory, the elements of the array can be accessed by their indices in constant time, simply by calculating the memory address of the element based on its index and the size of each element.

Contiguous blocks of memory are also used to store other types of data, such as strings, images, and other types of large data structures. Because the elements of a contiguous block of memory can be accessed sequentially, they can be processed more efficiently, for example by reading or writing the data in a loop.

# difference between array() and asarray()

### 1.5 the difference between array() and asarray()

#### array() and asarray() are two functions in the numpy library of Python that are used to create numpy arrays from other objects like lists, tuples, or other iterables. Here are the main differences between the two functions:

#### *array() function always creates a new numpy array and copies the data from the input, 
#### *asarray() function may return a reference to the same input array if it is already a numpy array.

# Example for array() function

```python
import numpy as np

# Using array() function
x = [1, 2, 3, 4]
y = np.array(x)
z = np.array(y) # creates a new copy of the array y

print(x)
print(y)
print(z)
print(id(y))
print(id(z))
```

###### In this example, we first create a Python list x and then create a numpy array y using the array() function. We then create a new numpy array z using array() again, this time with y as the input. The output shows that both y and z contain the same values as x, but z is a new copy of y.

# Example for asarray() function

```python
import numpy as np

# Using asarray() function
a = [1, 2, 3, 4]
b = np.asarray(a)
c = np.asarray(b) # returns the same array b

print(a)
print(b)
print(c)
print(id(b))
print(id(c))
```

1. The Python list a is defined as [1, 2, 3, 4].
2. The np.asarray() function is used to convert a to a NumPy array b.
3. b is then assigned to c, which creates a reference to the same array in memory.
4. The original list a, the NumPy array b, and the NumPy array c are printed using the print() function.

```python
a = np.array([1, 2, 3, 4, 5])
print(a[1:4])  
print("-----------------------")
b = np.array([[1, 2, 3], [4, 5, 6]])
print(b)
#print(#b[rows,column])
print(b[:, 1]) 
print(b[0, :2])  
```
